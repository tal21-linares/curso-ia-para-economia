```python
indicadores = {
    "crecimiento_pib":"NY.GDP.MKTP.KD.ZG",
    "pib_per_capita_ppa":"NY.GDP.PCAP.PP.KD",
    "inflacion":"FP.CPI.TOTL.ZG",
    "desempleo":"SL.UEM.TOTL.ZS",
    "poblacion":"SP.POP.TOTL"
}
```

```python
dataframes = []

for nombre, codigo in indicadores.items():

    url = f"https://api.worldbank.org/v2/country/DEU/indicator/{codigo}?date=2010:2023&format=json"

    response = requests.get(url)
    data = response.json()[1]

    df_temp = pd.DataFrame([{
        "anio": r["date"],
        nombre: r["value"]
    } for r in data])

    dataframes.append(df_temp)
```

```python
alemania_df = dataframes[0]

for df in dataframes[1:]:
    alemania_df = alemania_df.merge(df, on="anio")
```

```python
alemania_df.head()
```
